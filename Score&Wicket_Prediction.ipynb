```python
from sklearn import preprocessing
import keras 
import tensorflow as tf
```

```python
X = df.drop(['total_runs','batsman_runs','extra_runs','non_boundary','is_wicket'], axis =1)
y_run = df['total_runs']
y_wicket = df['is_wicket']
```

```python
X
```

```python
y_run
```

```python
y_wicket
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_run_train, X_run_test, y_run_train, y_run_test = train_test_split(X, y_run, test_size=0.3, random_state=42)
```

```python
X_wicket_train, X_wicket_test, y_wicket_train, y_wicket_test = train_test_split(X, y_wicket, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler 
scaler = MinMaxScaler()
```

```python
X_run_train_scaled = scaler.fit_transform(X_run_train)
X_run_test_scaled = scaler.transform(X_run_test)
```

```python
X_wicket_train_scaled = scaler.fit_transform(X_wicket_train)
X_wicket_test_scaled = scaler.transform(X_wicket_test)
```

```python
model1 = keras.Sequential([
    keras.layers.Input( shape=(X_run_train_scaled.shape[1],)),  
    keras.layers.Dense(512, activation='relu'), 
    keras.layers.Dense(216, activation='relu'),  
    keras.layers.Dense(1, activation='linear')  
])

huber_loss = tf.keras.losses.Huber(delta=1.0)  
model1.compile(optimizer='adam', loss=huber_loss)
```

```python
model1.fit(X_run_train_scaled, y_run_train, epochs=10, batch_size=64, validation_data=(X_run_test_scaled, y_run_test))
```

```python
model_losses = pd.DataFrame(model1.history.history)
model_losses.plot()
```

```python
predictions = model1.predict(X_run_test_scaled)
 
from sklearn.metrics import mean_absolute_error,mean_squared_error
mean_absolute_error(y_run_test,predictions)
```

```python
model2 = keras.Sequential([
    keras.layers.Input( shape=(X_run_train_scaled.shape[1],)),  
    keras.layers.Dense(512, activation='relu'), 
    keras.layers.Dense(216, activation='relu'),  
    keras.layers.Dense(1, activation='sigmoid')  
])
```

```python
model2.compile(loss='binary_crossentropy',
    optimizer='adam')
```

```python
model2.fit(X_wicket_train_scaled, y_wicket_train, epochs=10, batch_size=64, validation_data=(X_wicket_test_scaled, y_wicket_test))
```

```python
predictions_wicket = model2.predict(X_wicket_test_scaled)
 
from sklearn.metrics import mean_absolute_error,mean_squared_error
mean_absolute_error(y_wicket_test,predictions_wicket)
```
